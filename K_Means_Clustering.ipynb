```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('/home-rental.csv')
df.head()
```

```python
Kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster'] = Kmeans.fit_predict(df[['rent_amount','size']])
```

```python
centroids = Kmeans.cluster_centers_
```

```python
plt.scatter(df['rent_amount'],df["size"],c=df['Cluster'],cmap='rainbow')
plt.scatter(centroids[:,0],centroids[:,1],color='black', marker='x',s=200)
plt.xlabel('Rent Amount')
plt.ylabel('Size')
plt.title("K-means clustering example")
plt.show()
```
